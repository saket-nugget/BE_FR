### Foreround Only (Background Removed)

```python
from rembg import remove
from PIL import Image

# Load the image
input_path = "IMG_3591.png"
output_path = "output_no_bg_rembg.png"

input_image = Image.open(input_path)

# Remove background
output_image = remove(input_image)

# Save the image with transparent background
output_image.save(output_path)

print(f"Background removed and saved to {output_path}")

# Display the output image to confirm
output_image.show()
```

### Background Only (Foreground Removed)

```python
from rembg import remove
from PIL import Image, ImageOps

# Define paths
input_path = "IMG_3591.png"
output_path_background_only = "output_background_only.png"

# Load the original image, ensuring it has an alpha channel for manipulation
original_image = Image.open(input_path).convert("RGBA")

# 1. Use rembg to get the foreground (object) with transparency.
# The alpha channel of 'foreground_with_transparency' will be opaque for the foreground
# and transparent for the background regions.
foreground_with_transparency = remove(original_image)

# 2. Extract the alpha channel from the image where the foreground is present.
alpha_channel_fg = foreground_with_transparency.getchannel('A')

# 3. Invert this alpha channel.
# Now, the areas that were originally background will be opaque (255),
# and the areas that were originally foreground will be transparent (0).
inverted_alpha_channel = ImageOps.invert(alpha_channel_fg)

# 4. Create a new image using the original RGB data and the inverted alpha channel.
# This effectively makes the foreground transparent while keeping the background intact.
original_rgb = original_image.convert("RGB")
background_only_image = original_rgb.copy()
background_only_image.putalpha(inverted_alpha_channel)

# Save the resulting image
background_only_image.save(output_path_background_only)

print(f"Background retained (foreground removed) and saved to {output_path_background_only}")

# Display the output image to confirm
background_only_image.show()
```
